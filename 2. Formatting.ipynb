# Formatting

Graphs are more appealing if they look better. Lets see what we can do



```python
import matplotlib.pyplot as plt
%matplotlib inline
```

Let me use the same data as seen in the 1st of the series

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

## Markers

[Doc link](https://matplotlib.org/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D.markers)
This creates the required shape at the coordinates plotted

```python
plt.plot(x, y, marker = 'd')
```

```python
plt.plot(x, y, marker = '1', markersize= '15', linestyle = '')
```

## Transparency

Blending or transperancy can be set with alpha value. Its [Doc link](https://matplotlib.org/api/_as_gen/matplotlib.artist.Artist.set_alpha.html#matplotlib.artist.Artist.set_alpha)

```python
plt.plot(x, y, marker = '*', linestyle = '--', alpha = 0.4)
```

## Multiple Charts

Pretty much the same thing. You must provide the plot function multiple times for every data



```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)
```

## Adding a legend

Just to make it a bit informative. Give each plot a label to be shown in the legend. Also, call the legend using the **legend** function

```python
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")

plt.legend()
```

A lot can be done with legends. Refer these [Docs](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.legend.html?highlight=legend#matplotlib-pyplot-legend)

## Grids

For relative measurements, grids are pretty awesome.

```python
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")

plt.legend()
plt.grid()
```
