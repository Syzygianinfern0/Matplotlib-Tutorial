  You can generate plots, histograms, power spectra, bar charts, errorcharts, scatterplots, etc., with just a few lines of code. [Check this Out](https://matplotlib.org/tutorials/introductory/sample_plots.html)

### Plotting basics

The following data is to be plotted

| Day | Temperature|
| :-------------: |:-------------:|
| 1 | 50 |
| 2 | 51 |
| 3 | 52 |
| 4 | 48 |
| 5 | 47 |
| 6 | 49 |
| 7 | 46 |

````python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
plt.plot(x, y)
````

### Customization

Provide a **colour** arguement to change it colour.

````python
plt.plot(x, y, color = 'green')
````

A **linewidth** and **linestyle** arguements are possible too. 

````python
plt.plot(x, y, color = 'black', linewidth = 5, linestyle = '-.')
````

Cool right? Check out their [Docs](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib-pyplot-plot) for all you will need is there about customizing your plot. 

### Naming

X lables, Y labels and titles are the most commonly used labels

````python
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.title('Weather Chart')
plt.plot(x, y, color = 'red', linewidth = 2, linestyle = ':')
````
